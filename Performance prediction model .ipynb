```python
inputDF = pd.DataFrame(np.c_[df['TLR'], df['RPC'], df['GO'], df['OI'], df['Perception']], columns = ['TLR','RPC','GO','OI','Perception'])
y = df['Score'].values
                                                             
xTrain, xTest, yTrain, yTest = train_test_split(inputDF, df['Score'], test_size = 0.2, random_state = 0)   
```

```python
from sklearn import linear_model
modelMLR = linear_model.LinearRegression().fit(xTrain, yTrain)
print('Multiple Linear Regression \n')
print('R2 Score:', modelMLR.score(xTest, yTest))

y_pred = modelMLR.predict(xTest)
```

```python
#lasso or L1 regularization
from sklearn.linear_model import Lasso
model2 = Lasso().fit(xTrain, yTrain)
model2.score(xTest,yTest)
```

```python
#Ridge or L2 regularization
from sklearn.linear_model import Ridge
model2 = Ridge().fit(xTrain, yTrain)
model2.score(xTest,yTest)
```

```python
features = ['TLR','RPC','GO','OI','Perception']
```

```python
from sklearn.tree import DecisionTreeRegressor 
min_samples = [1,5,10,14,16]

for i in min_samples:
    regressor = DecisionTreeRegressor(min_samples_leaf = i)
    regressor.fit(xTrain,yTrain)
    predicted_labels = regressor.predict(xTest)
    accuracyDT = regressor.score(xTest, yTest)
    print("\n Accuracy with min samples at leaf node:", i,"-",accuracyDT, '\n')
```

```python
from sklearn import neighbors
error_rate =[]
for i in range(2,20):
    regressor = neighbors.KNeighborsRegressor(n_neighbors = i)
    regressor.fit(xTrain,yTrain)
    predicted_labels = regressor.predict(xTest)
    expected_labels = list(yTest)
    error_rate.append((np.square(np.subtract(predicted_labels, expected_labels))).mean())
```

```python
sns.lineplot(range(2,20), error_rate)
plt.title('Error rate vs K')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
plt.savefig('error_vs_k.eps')
```

```python
regressorKNN = neighbors.KNeighborsRegressor(n_neighbors = 4)
regressorKNN.fit(xTrain,yTrain)
predicted_labels = regressorKNN.predict(xTest)
accuracyKNN = regressorKNN.score(xTest, yTest)
print("Accuracy:", accuracyKNN)
```

```python
from sklearn.ensemble import RandomForestRegressor

for i in range(4,13,2):
    regressor = RandomForestRegressor(max_depth = i)
    regressor.fit(xTrain,yTrain)
    predicted_labels = regressor.predict(xTest)
    accuracyRF = regressor.score(xTest, yTest)
    print("\n Accuracy for max-depth:", i,":",accuracyRF, '\n')
```

```python
regressorRF = RandomForestRegressor(max_depth = 10)

regressorRF.fit(xTrain,yTrain)
accuracyRF = regressorRF.score(xTest, yTest)
print("Accuracy:", accuracyRF)
```

```python
#visualisation of random forest with max depth of the trees =3(5 of the decision trees are displayed here)
from sklearn import tree
regressorNo = 0

regressor = RandomForestRegressor(max_depth = 3, random_state =0)
regressor.fit(xTrain,yTrain)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (10,2), dpi =1000)
for i in range(0,2):
    for j in range(0,2):
        tree.plot_tree(regressor.estimators_[regressorNo],
                      feature_names = features,
                      filled = True,
                      ax =axes[i][j]);
        regressorNo += 1
        
    fig.savefig('4trees.png')
```

```python
from sklearn.cluster import KMeans

#creating 200 clusters equivalent to 200 ranks

kmeans_model = KMeans(n_clusters = 200)
df_parameters['clusters'] = kmeans_model.fit_predict(df_parameters)
```

```python
 def add_region(df):
        if df.State in ['Himachal Pradesh','Punjab','Uttarakhand','Uttar Pradesh','Haryana','Delhi','Chandigarh','Jammu and Kashmir']:
            return 'North'
        elif df.State in ['Bihar','Odisha','Jharkhand','West Bengal','Assam','Sikkim','Nagaland','Meghalaya','Manipur','Mizoram']:
            return 'East'
        elif df.State in ['Rajasthan','Gujarat','Goa','Maharashtra','Madhya Pradesh','Chattisgarh']:
            return 'West'
        elif df.State in ['Andhra Pradesh','Karnataka','Kerala','Tamil Nadu','Telangana','Pondicherry','Nagaland','Andaman and Nicobar Islands']:
            return 'South'
        
        else:
            return 'Unkonwn'
```

```python
dataframes = [data2016,data2017,data2018,data2019,data2020]
```

```python
df_region_mean = []

for dataframe in dataframes:
    dataframe['region'] = (dataframe.apply(lambda dataframe: add_region(dataframe), axis =1))
    df_region_mean.append(dataframe.groupby('region').mean())
    

df_region_mean[0]
```
